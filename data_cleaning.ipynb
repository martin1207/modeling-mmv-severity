```python
import pandas as pd
import geopandas as gpd
```

```python
accidents_lyon=gpd.read_file('acci_lyon.geojson')
accidents_paris=gpd.read_file('acci_paris.geojson')
data_lyon=gpd.read_file('data_lyon.gpkg')
data_paris=gpd.read_file('data_paris.gpkg')
```

```python
paris=pd.merge(data_paris,accidents_paris.drop(columns=['int','geometry']),on='Num_Acc_x',how='left')
lyon=pd.merge(data_lyon,accidents_lyon.drop(columns=['int','geometry']),on='Num_Acc_x',how='left')
paris.rename(columns={'Num_Acc_x':'Num_Acc'},inplace=True)
lyon.rename(columns={'Num_Acc_x':'Num_Acc'},inplace=True)
paris=paris.rename(columns={'EPCI_name_u':'ville'})
```

```python
# Combine the data from Paris and Lyon into a single DataFrame
acci = pd.concat([paris, lyon])
```

```python
"""
Cell generated by Data Wrangler.
"""
def clean_data(acci):
    # Drop column: 'dep'
    acci = acci.drop(columns=['dep'])
    # Drop column: 'agg'
    acci = acci.drop(columns=['agg'])
    # Drop column: 'occutc'
    acci = acci.drop(columns=['occutc'])


    # Drop column: 'secu3'
    acci = acci.drop(columns=['secu3'])
    # Drop column: 'v2'
    acci = acci.drop(columns=['v2'])
    # Drop column: 'v1'
    acci = acci.drop(columns=['v1'])
    # Drop column: 'pr1'
    acci = acci.drop(columns=['pr1'])
    # Drop column: 'pr'
    acci = acci.drop(columns=['pr'])
    # Drop column: 'lartpc'
    acci = acci.drop(columns=['lartpc'])
    # Drop column: 'larrout'
    acci = acci.drop(columns=['larrout'])
    return acci

acci_clean = clean_data(acci.copy())
acci_clean.head()
```

```python
"""
Cell generated by Data Wrangler.
"""
def clean_data(acci_clean_1):
    # Replace all instances of "Sens Unique" with "UNIQUE" in column: 'senscircul'
    acci_clean_1['senscirculation'] = acci_clean_1['senscirculation'].str.replace("Sens Unique", "UNIQUE", case=False, regex=False)
    # Replace all instances of "double" with "DOUBLE" in column: 'senscircul'
    acci_clean_1['senscirculation'] = acci_clean_1['senscirculation'].str.replace("double", "DOUBLE", case=False, regex=False)
    acci_clean_1['senscirculation'].fillna(acci_clean_1['sens_voit'],inplace=True)
    # Drop column: 'sens'
    acci_clean_1 = acci_clean_1.drop(columns=['sens'])
    return acci_clean_1

acci_clean_2 = clean_data(acci_clean_1.copy())
```

```python
"""
Cell generated by Data Wrangler.
"""
def clean_data(acci_clean_3):
    # Drop column: 'codetronco'
    acci_clean_3 = acci_clean_3.drop(columns=['codetronco'])
    # Drop column: 'geo_point_2d'
    acci_clean_3 = acci_clean_3.drop(columns=['geo_point_2d'])
    # Drop column: 'tram_crossing'
    acci_clean_3 = acci_clean_3.drop(columns=['tram_crossing'])
    # Drop column: 'hierarchi0'
    acci_clean_3 = acci_clean_3.drop(columns=['hierarchi0'])
    # Drop column: 'revetemen0'
    acci_clean_3 = acci_clean_3.drop(columns=['revetemen0'])
    # Drop column: 'revetemen1'
    acci_clean_3 = acci_clean_3.drop(columns=['revetemen1'])
    # Drop column: 'senscirculation'
    acci_clean_3 = acci_clean_3.drop(columns=['senscirculation'])
    # Drop column: 'limitation'
    acci_clean_3 = acci_clean_3.drop(columns=['limitation'])
    # Replace all instances of "z30" with "Zone 30" in column: 'nv'
    acci_clean_3['nv'] = acci_clean_3['nv'].str.replace("z30", "Zone 30", case=False, regex=False)
    # Replace all instances of "z20" with "Zone de rencontre" in column: 'nv'
    acci_clean_3['nv'] = acci_clean_3['nv'].str.replace("z20", "Zone de rencontre", case=False, regex=False)
    # Replace all instances of "rue pietonne" with "Aire Piétonne" in column: 'nv'
    acci_clean_3['nv'] = acci_clean_3['nv'].str.replace("rue pietonne", "Aire Piétonne", case=False, regex=False)
    acci_clean_3['reglementa'].fillna(acci_clean_3['nv'],inplace=True)
    # Drop column: 'insee_com'
    acci_clean_3 = acci_clean_3.drop(columns=['insee_com'])
    # Drop column: 'nv'
    acci_clean_3 = acci_clean_3.drop(columns=['nv'])
    # Rename column 'longueurtr' to 'longueur_trottoir_droit'
    acci_clean_3 = acci_clean_3.rename(columns={'longueurtr': 'longueur_trottoir_droit'})
    # Rename column 'surfacetro' to 'surface_trottoir_droit'
    acci_clean_3 = acci_clean_3.rename(columns={'surfacetro': 'surface_trottoir_droit'})
    # Rename column 'largeurtr0' to 'largeur_trottoir_droit'
    acci_clean_3 = acci_clean_3.rename(columns={'largeurtr0': 'largeur_trottoir_droit'})
    # Rename column 'longueurt0' to 'longueur_trottoir_gauche'
    acci_clean_3 = acci_clean_3.rename(columns={'longueurt0': 'longueur_trottoir_gauche'})
    # Rename column 'surfacetr0' to 'surface_trottoir_gauche'
    acci_clean_3 = acci_clean_3.rename(columns={'surfacetr0': 'surface_trottoir_gauche'})
    return acci_clean_3

acci_clean_4 = clean_data(acci_clean_3.copy())
acci_clean_4.head()
```

```python
def determine_season(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    elif month in [9, 10, 11]:
        return 'Autumn'
    else:
        return 'Unknown'  # Au cas où il y aurait une valeur invalide

# Appliquer la fonction pour créer la colonne 'season'
acci_clean_7['season'] = acci_clean_7['mois'].apply(determine_season)
```
